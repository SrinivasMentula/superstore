```python
%sql 
--creating the  required unity catlog and database  opeartions 
create catalog if not exists superstore;
create database if not exists superstore.sales_reporting;
use catalog superstore;
use database sales_reporting;
CREATE VOLUME IF NOT EXISTS superstore.sales_reporting.Inbound;
```

```python
%sql
/**Table creation for the execution logs**/
create table if not exists superstore.sales_reporting.executionLog (
    logId BIGINT generated always AS identity,
    noteBookName varchar(100),
    status varchar(100),
    createAt timestamp default current_timestamp,
    createBy string
  ) using delta
  tblproperties ('delta.feature.allowColumnDefaults' = 'supported')
```

```python
%sql
/**Tbale creation for the error logs**/
CREATE TABLE IF NOT EXISTS superstore.sales_reporting.exceptionLog(
    exceptionId  BIGINT GENERATED  ALWAYS as IDENTITY , 
    notebookName  varchar(100),
    error_message  STRING, 
    createdAt timestamp default current_timestamp
)
using delta 
tblproperties(
  'delta.feature.allowColumnDefaults' = 'supported'
)
```
